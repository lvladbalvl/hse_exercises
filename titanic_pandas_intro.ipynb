### Работа с NaN значениями

```python
df['Age'].fillna(df['Age'].median(),inplace=True)
```

```python
df['Age'].isnull().any()
```

```python
df.dropna(subset=['Age'],inplace=True)
```

### Разделение на бакеты

```python
pd.cut(df["Age"],bins=10)
```

```python
df["CAge"]=pd.cut(df["Age"], bins = [0,10,18,40,max(df["Age"])] ,labels=["Child","Adolescent","Young adult","Adult"])
```

```python
pd.crosstab(df['CAge'],df['Survived'])
```

```python
sns.boxplot(x='Survived', y='Fare', data=df)
```

Распределение цены за билет логнормальное - можно создать дополнителный столбец прологорифмировав

```python
df['logFare'] = df['Fare'].apply(np.log1p)
```

```python
sns.boxplot(x='Survived', y='logFare', data=df)
```

```python
plt.hist(df[df['Survived']==0]['Fare'],bins=20)
plt.hist(df[df['Survived']==1]['Fare'],bins=20);
plt.legend(['Died','Survived'])
plt.xlabel('Fare')
```

```python
plt.hist(df[df['Survived']==0]['Fare'],bins=50)
plt.hist(df[df['Survived']==1]['Fare'],bins=50,alpha=0.5);
plt.legend(['Died','Survived'])
plt.xlabel('Fare')
```

```python
plt.hist(df[df['Survived']==0]['logFare'],bins=50)
plt.hist(df[df['Survived']==1]['logFare'],bins=50,alpha=0.5);
plt.legend(['Died','Survived'])
plt.xlabel('logFare')
```

```python
sns.distplot(df[df['Survived']==0]['logFare'],bins=50)
sns.distplot(df[df['Survived']==1]['logFare'],bins=50);
plt.legend(['Died','Survived'])
plt.xlabel('logFare')
```

```python
sns.boxplot(x='Pclass', y='logFare', data=df)
```

```python
sns.distplot(df[df['Pclass']==1]['logFare'],bins=50)
sns.distplot(df[df['Pclass']==2]['logFare'],bins=50)
sns.distplot(df[df['Pclass']==3]['logFare'],bins=50)
plt.legend(['First','Second','Third'])
plt.xlabel('logFare')
```
